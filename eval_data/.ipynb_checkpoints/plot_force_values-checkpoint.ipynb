```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=[]
element =[0,[0]]

with open('logfile.txt') as f:
    line = f.readline()
    while line:
        if (line.find("1") == 0):
            element[0]=int(line.strip())
            line = f.readline()
            if (line.find("[") == 0):
                stripped_line = line[1:].strip()
                line = f.readline()
                while (line.find("]") == -1):
                    stripped_line += " "+line.strip()
                    line = f.readline()
                if (line.find("]") > 0):
                    stripped_line += " "+line[:-2].strip()
                    element[1] = [float(k) for k in stripped_line.split()]
                    data.append(element.copy())
                    line = f.readline()
            
                
```

```python
datanr = len(data)
tendonnr = len(data[1][1])
print("Data Points: %i \nTendons: %i" % (datanr,tendonnr))
```

```python
#start at specific index?
startindex = 30
data = data[startindex:]
```

```python
datanr = len(data)
tendonnr = len(data[1][1])
print("Data Points: %i \nTendons: %i" % (datanr,tendonnr))
```

```python
fig = plt.gcf()
fig.set_size_inches(20, 16)


x = [m[0] for m in data]

#start x axis with 0
tmin = min([m[0] for m in data])
x = [t - tmin for t in x]


for r in range(8,17):
    plt.plot(x,[n[1][r] for n in data])


plt.ylabel('force')
plt.xlabel('time')

plt.show
```
